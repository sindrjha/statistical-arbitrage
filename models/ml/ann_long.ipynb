```python
from joblib import Parallel, delayed

# Define parameters for validation, test, and window sizes
validation_size = 250
test_size = 250
window_size = 5

# List of hub pairs and parameter dictionary
pairs = ['ttf-the', 'ttf-nbp', 'the-nbp']

# Define the processing function
def process_pair(pair):
    hub1_name, hub2_name = pair.split('-')
    print(f'Processing {hub1_name} and {hub2_name}')
    params = param_dict[pair]
    ann = ann_system_rolling(hub1_name, hub2_name, validation_size, test_size, window_size, params, verbose=True, save=True)
    return ann

# Run the function in parallel for each pair
if __name__ == "__main__":
    results = Parallel(n_jobs=3)(delayed(process_pair)(pair) for pair in pairs)
```
